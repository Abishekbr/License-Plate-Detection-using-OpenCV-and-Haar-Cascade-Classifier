```
import cv2
import matplotlib.pyplot as plt
```

## Load and Display Image

```
img = cv2.imread('car.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
```

## Convert to Grayscale

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

## Preprocessing (Improvement)

```
blur = cv2.GaussianBlur(gray, (5,5), 0)
equalized = cv2.equalizeHist(blur)

plt.imshow(equalized, cmap='gray')
plt.title("Preprocessed Image")
plt.axis('off')
```

## License Plate Detection

```
plate_cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')

plates = plate_cascade.detectMultiScale(equalized, 1.1, 5)

print("Number of plates detected:", len(plates))
```

## Draw Bounding Boxes

```
output = img.copy()

for (x, y, w, h) in plates:
    cv2.rectangle(output, (x, y), (x+w, y+h), (0,255,0), 2)

plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title("Detected Plates")
plt.axis('off')
```

## Crop and Save Plates

```
count = 0
for (x, y, w, h) in plates:
    plate = img[y:y+h, x:x+w]
    cv2.imwrite(f'plate_{count}.png', plate)
    count += 1

print("Cropped plates saved.")
```

## Conclusion
Haar Cascade successfully detects license plates. Preprocessing improves accuracy. Blob-like detection is sensitive to parameters such as scaleFactor and minNeighbors.
